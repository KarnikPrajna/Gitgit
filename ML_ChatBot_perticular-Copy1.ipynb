```python
#function to return a random greeting response to a user's greeting
def greet_response(text):
    val = text.lower()
    user_greet = "hey,hi,hello,howdy,hola,greetings"
    bot_greet = ['Halo', 'Hey', 'Greetings', 'Hola', 'Wassup', 'Hi', 'Hello']
    
    for word in val.split():
        if word in user_greet.split(","):
            return random.choice(bot_greet)
```

```python
def bank(inp):
    use_inp = ["what is your name", "what is your age", "i am bored", "what is on the menu", "which is your favourite football team"]
    bot_out = ["you can call me Doc", "i am 20 yours young", "let us buy cookies together", "cookies are on the menu", "your favourite is my favourite"]
    inpu = inp.lower()
    for i in range(len(use_inp)):
        if use_inp[i] == inpu :
            return(bot_out[i])
```

```python
def index_sort(list_var):
    length = len(list_var)
    list_index = list(range(0, length))
    
    x = list_var
    for i in range(length):
        for j in range(length):
            if x[list_index[i]] > x[list_index[j]]:
                #sort: user input which matches the most is kept at the highest prioruty
                temp = list_index[i]
                list_index[i] = list_index[j]
                list_index[j] = temp
                
    return list_index
```

```python
# function to respond- by bot
def bot_response(user_input):
    user_input = user_input.lower()
    senlist.append(user_input)
    bot_response = ''
    cm = CountVectorizer().fit_transform(senlist)
    similar_scores = cosine_similarity(cm[-1], cm)  # check for similarity of user's last sentence & all d sentences in senlist
    similar_scores_list = similar_scores.flatten()  # similar_scores_ist contaon a range of values
    index = index_sort(similar_scores_list)  # index will contain the values of indeces of highest values & similarity scores
    index = index[1:]
    response_flag = 0
    j = 0  # this helps to return only top two similar sentences to the user
    for i in range(len(index)):
        if similar_scores_list[index[i]] > 0.0:
            bot_response = bot_response+' '+senlist[index[i]]
            response_flag = 1
            j = j+1
        if j>2:
            break
    if response_flag == 0:
        bot_response = bot_response+' '+"Sorry, I dont understand"
    
    senlist.remove(user_input)
    
    return(bot_response)
```
